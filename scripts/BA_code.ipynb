## Replace Values
https://datatofish.com/replace-values-pandas-dataframe/

`df['column name'] = df['column name'].replace(['old value'],'new value')`

```python
#change characters that werent displayed right
df = df.replace('Ã¤', 'ä', regex=True)
df = df.replace('Ã¶', 'ö', regex=True)
df = df.replace('Ã¼', 'ü', regex=True)
df = df.replace('ÃŸ', 'ß', regex=True)
df = df.replace('Ã©', 'é', regex=True)


df['quest_correct_key'] = df['quest_correct_key'].replace([1, 0], ['up', 'down'])

# change handedness to a uniform 'l' and 'r'
df['handedness'] = df['handedness'].replace(['rechts', 'Rechts','Rechtshänder','Rechtshänderin'], 'r')
df['handedness'] = df['handedness'].replace(['linkshänder', 'Linkshänder'], 'l')
df['participant_gender'] = df['participant_gender'].replace(['weiblich', 'w'], 'w')
df['participant_gender'] = df['participant_gender'].replace(['w/nicht-binär', 'nicht-binär'], 'nb')
df['participant_gender'] = df['participant_gender'].replace(['kein Angabe'], 'NA')

print(df['quest_correct_key'])
print('--------------------------------------------------------------------------------------')
print(df['handedness'])
```
